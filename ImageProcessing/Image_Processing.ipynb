```python
from skimage import io
```

```python
# Loading
img1 = io.imread('bird.jpg')
img1.shape
```

```python
io.imshow(img1)
```

```python
from skimage import color

gray_image = color.rgb2gray(img1)
gray_image.shape
```

```python
io.imshow(gray_image)
```

```python
hsv_img = color.rgb2hsv(img1)
io.imshow(hsv_img)
```

```python
from skimage import transform
img3 = transform.rescale(img1, 1/2, multichannel=True)
img3.shape
io.imshow(img3)
```

```python
img3 = transform.rescale(img1, 3.0, multichannel=True)
img3.shape
io.imshow(img3)
```

```python
img3.shape
```

```python
# Resizing
res_img = transform.resize(img1, (64,100))
io.imshow(res_img)
```

```python
io.imsave("bird1.jpg", res_img)
```

```python
from skimage import exposure
exp_image = exposure.adjust_gamma(img1, 0.2)
io.imshow(exp_image)
```

```python
exp2 = exposure.adjust_sigmoid(img1, cutoff=0.6)
io.imshow(exp2)
```
